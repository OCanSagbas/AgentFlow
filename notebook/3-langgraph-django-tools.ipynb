```python
from django.contrib.auth import get_user_model
User = get_user_model()

# Tüm kullanıcıları detaylı listele
for user in User.objects.all():
    print(f"\n{'='*50}")
    print(f"User ID: {user.id}")
    print(f"Username: {user.username}")
    print(f"Email: {user.email}")
    print(f"Is Staff: {user.is_staff}")  # Admin paneline giriş
    print(f"Is Superuser: {user.is_superuser}")  # Tüm yetkiler
    print(f"Is Active: {user.is_active}")
    
    # Bu kullanıcının document sayısı
    from documents.models import Document
    doc_count = Document.objects.filter(owner=user).count()
    print(f"Document Count: {doc_count}")
    
    # Groups (varsa)
    groups = user.groups.all()
    print(f"Groups: {[g.name for g in groups]}")
```
